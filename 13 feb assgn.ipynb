```python
#Q2. Write a python program to print Python Exception Hierarchy.
"""Sure, here's an example Python program that prints the Python exception hierarchy using the __subclasses__() method:
"""
class_hierarchy = []
for cls in Exception.__subclasses__():
    class_hierarchy.append(cls.__name__)
    for sub_cls in cls.__subclasses__():
        class_hierarchy.append("  " + sub_cls.__name__)

print("\n".join(class_hierarchy))
"""This program creates an empty list called class_hierarchy, and then iterates over the direct subclasses of the Exception class using the __subclasses__() method. For each subclass, it appends the class name to the class_hierarchy list, and then iterates over the subclasses of that subclass, adding each subclass name to the list with two leading spaces to indicate the hierarchical relationship.

Finally, the program uses the join() method to join the elements of the class_hierarchy list with newline characters, and prints the resulting string"""
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
"""The LookupError class is a built-in Python exception class that serves as a base class for several other specific exception classes related to lookup errors. This class is a subclass of the Exception class and is typically used when an index or key lookup fails.

Here are two examples of specific lookup errors that are subclasses of the LookupError class:

KeyError: This error occurs when you try to access a dictionary key that does not exist. For example:
"""
my_dict = {"apple": 2, "banana": 3, "orange": 5}

try:
    value = my_dict["grape"]
except KeyError:
    print("Error: Key not found")
"""In this example, the try block attempts to access the value associated with the key "grape" in the my_dict dictionary. Since this key does not exist in the dictionary, a KeyError is raised, which is caught by the except block and the error message "Error: Key not found" is printed.

IndexError: This error occurs when you try to access a list or tuple index that is out of range. For example:
"""
my_list = [1, 2, 3, 4, 5]

try:
    value = my_list[10]
except IndexError:
    print("Error: Index out of range")
"""In this example, the try block attempts to access the value at index 10 in the my_list list. However, since the list only has five elements, this index is out of range, causing an IndexError to be raised. This error is caught by the except block and the error message "Error: Index out of range" is printed.

In summary, the LookupError class is a base class for several specific lookup errors that can occur during index or key lookups. Two examples of lookup errors that are subclasses of the LookupError class are KeyError, which occurs when attempting to access a dictionary key that does not exist, and IndexError, which occurs when attempting to access a list or tuple index that is out of range"""
```
